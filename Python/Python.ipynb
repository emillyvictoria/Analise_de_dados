#Aplicando conceitos



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('framingham.csv')
df.head()
```

```python
plt.hist(df['heartRate'])
```

```python
plt.figure (figsize= (9,6))
plt.hist (df['heartRate'], bins = 20, color = 'yellow') #bins: quant de categorias ou espaçamentos de um gráfico
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.scatter(df['diaBP'], df['sysBP'],color='red')
plt.xlabel('diaBP')
plt.ylabel('sysBP')
plt.title('diaBP X sysBP')
plt.show()
```

```python
df.corr()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols = 2, nrows = 1, figsize = (18,6))#aparecer dois gráficos

ax1.hist (df['heartRate'], bins = 20, color = 'yellow')
plt.grid()

ax2.scatter(df['diaBP'], df['sysBP'],color='red')
ax2.grid()
```

#Seaborn

```python
plt.figure(figsize=(9,6))
sns.distplot(df['heartRate'], bins= 20) #histograma com linha de tendência
plt.show()
```

```python
plt.figure (figsize= (9,6))
sns.scatterplot(data = df, x='diaBP', y= 'sysBP', hue= 'prevalentHyp', palette= ['yellow', 'pink'])
plt.grid()
plt.show()
#traduzindo: em pessoas com 'prevalentHyp' é mais provável que elas tenham a 'diaBP' maior 
```

```python
sns.jointplot(data= df, x='diaBP', y='sysBP', height= 7, color ='purple') #faz um scatter misturado com um histograma
plt.show()
```

```python
plt.figure(figsize=(9,6))
sns.boxplot(data = df, x = 'diabetes', y = 'glucose', palette=['yellow', 'pink'])
plt.show()
#é traçado um eixo x uma ver categorica e no eixo y uma var numérica
#serve para comparar como as categorias se comportam em relação à variável numérica (se existe alguma diferença)
#a linha preta do meio significa a média 
```

```python
plt.figure(figsize=(9,6))
sns.violinplot(data = df, x = 'diabetes', y = 'glucose', palette=['yellow', 'pink'])
plt.show()
#parecido com o boxplot k
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(), annot=True)
plt.show()
#mostra um gráfico de um mapa de calor entre as correlações no gráfico
#é possível ver as variáveis com maior correlação através da barrinha com o guia para cores maus escuras e claras
```

```python
colunas = ['totChol', 'diaBP', 'sysBP', 'age']
sns.pairplot(data = df, vars = colunas)
#em apenas um gráfico é possível ver o histograma, scatter e seu relacionamenro com as outras categorias através de scatter também
```

```python
sns.pairplot(data = df, vars = colunas, hue = 'prevalentHyp')
#hue separa os dados
```
